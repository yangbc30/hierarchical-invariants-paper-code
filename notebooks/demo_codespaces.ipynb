# Photonic Jordan Quick Demo

This notebook is a minimal sanity-check that can run directly in GitHub Codespaces.

```python
import numpy as np
from photonic_jordan import PhotonicSystem

sys = PhotonicSystem(m_ext=2, n_particles=3, rng=np.random.default_rng(0))
rho = sys.state.from_modes_and_gram([0, 1, 0], gram=0.4)

print('sector weights:', rho.sector_weights())
print('I_exact(1, sector=(2,1)):', rho.invariant.I_exact(1, sector=(2, 1)))
print('I_exact(1, multiplicity=((2,1),0)):', rho.invariant.I_exact(1, multiplicity=((2, 1), 0)))
```

```python
S = sys.unitary.haar(seed=123)
rho_out = rho.evolve(S)

for j in range(3):
    lhs = rho.invariant.I_cumulative(j, sector=(2, 1))
    rhs = rho_out.invariant.I_cumulative(j, sector=(2, 1))
    print(f'j={j}: sector cumulative diff = {abs(lhs-rhs):.3e}')
```
